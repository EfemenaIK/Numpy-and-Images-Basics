## Image Processing using Numpy
Images can also be considered as arrays, we can use NumPy for performing different image processing tasks

```python
import numpy as np
import matplotlib.pyplot as plt

# PIL - Python Imaging Library (Pillow)
from PIL import Image as ig
```

```python
# Load the image
rot = ig.open(r'C:\Users\ikpro\OneDrive\Desktop\Deskstop\Rot.jpg')
rot
```

```python
# Checking the type of data that was brought in, you will observe it is an Image file (i.e. Jpeg file). Which means Numpy can't work with it. 
type(rot)
```

```python
# We will need to transform it into an array. 
rot_array = np.asarray(rot)
type(rot_array)
```

```python
rot_array.shape
```

Taking a look at the shape of "rot_array" will display the dimension of the picture (Height-380, Width-266 and 3 signifies the three color channels available (RGB)

```python
plt.imshow(rot_array)
```

```python
# Lets access the color channels available in the picture.
# We are going to start off with the RED color channel
rot_red = rot_array.copy()
rot_red.shape
```

```python
# Red = 0, Green = 1, Blue = 2
plt.imshow(rot_red[:,:,0])
```

```python
# For better understanding, We are going to visualise it in grayscale using cmap(Color Map) 
# The red channel is between 0 - 255 (255 means a full pure Red while 0 means no red). Indicating the closer to white more red
plt.imshow(rot_red[:,:,0], cmap='gray')
```

```python
# The Green channel is between 0 - 255
plt.imshow(rot_red[:,:,1], cmap='gray')
```

```python
# The blue channel is between 0 - 255
plt.imshow(rot_red[:,:,2], cmap='gray')
```

```python
# Let remove different color scales and observe the changes
# Lets take a look at the red color scale array
rot_red[:,:,0]
```

```python
# I am going to assign everything in that array to 0 and see what happens
rot_red[:,:,0] = 0
plt.imshow(rot_red)
```

```python
# Taking out red and green will leave only the blue color scale available
rot_red[:,:,1] = 0
plt.imshow(rot_red)
```
